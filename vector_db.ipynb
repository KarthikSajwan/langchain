```
from langchain_chroma import Chroma
from langchain_openai import OpenAIEmbeddings
```

```
from langchain_core.documents import Document
doc1 = Document(
    page_content="Virat Kohli is one of the most successful and consistent batsmen in IPL history. He is known for his aggressive batting style and has been a key player for RCB for many seasons.",
    metadata={"team": "Royal Challengers Bangalore", "player": "Virat Kohli"}
)

doc2 = Document(
    page_content="Rohit Sharma is the most successful captain in IPL history, leading Mumbai Indians to multiple titles. He is known for his calm leadership and elegant batting.",
    metadata={"team": "Mumbai Indians", "player": "Rohit Sharma"}
)

doc3 = Document(
    page_content="MS Dhoni, popularly known as Captain Cool, has led Chennai Super Kings to several IPL championships. He is famous for his finishing skills and sharp wicketkeeping.",
    metadata={"team": "Chennai Super Kings", "player": "MS Dhoni"}
)

doc4 = Document(
    page_content="Jasprit Bumrah is considered one of the best fast bowlers in T20 cricket. Playing for Mumbai Indians, he is known for his deadly yorkers and consistency.",
    metadata={"team": "Mumbai Indians", "player": "Jasprit Bumrah"}
)

doc5 = Document(
    page_content="Ravindra Jadeja is a world-class all-rounder who contributes with both bat and ball. Representing Chennai Super Kings, he is also one of the best fielders in the game.",
    metadata={"team": "Chennai Super Kings", "player": "Ravindra Jadeja"}
)
```

```
docs = [doc1, doc2, doc3, doc4, doc5]
```

```
import os
from google.colab import userdata

os.environ["OPENAI_API_KEY"] = userdata.get("OPENAI_API_KEY")
```

```
vector_store = Chroma(
    embedding_function=OpenAIEmbeddings(),
    persist_directory="chroma_db",
    collection_name="sample"
)
```

```
vector_store.add_documents(docs)
```

```
vector_store.get(include=['embeddings', 'documents', 'metadatas'])
```

```
vector_store.similarity_search(
    query="Who among these are bowlers?",
    k=2
)
```

```
vector_store.similarity_search_with_score(
    query="Who among these are bowlers?",
    k=2
)
```

```
vector_store.similarity_search_with_score(
    query="",
    filter={"team": "Mumbai Indians"}
)
```

```
update_doc1 = Document(
    page_content="Virat Kohli is one of the most successful and consistent batsmen in IPL history. He is known for his aggressive leadership and batting style and has been a key player for RCB for many seasons.",
    metadata={"team": "Royal Challengers Bangalore"}
)
vector_store.update_document(document_id="72985c3d-6412-414b-9471-7bf10f5fbd2a", document=update_doc1)
```

```
result = vector_store.get(
    ids=["72985c3d-6412-414b-9471-7bf10f5fbd2a"]
)

result
```
